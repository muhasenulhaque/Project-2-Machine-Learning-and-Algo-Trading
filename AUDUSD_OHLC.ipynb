```python
# Import dotenv package for setting environment variables 
from dotenv import load_dotenv

# Import os package
import os
```

```python
import pandas as pd
from pathlib import Path
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()
```

```python
# Retrieve API key and store as Python variable
api_key = os.getenv("TRADERMADE_API_KEY")
type(api_key)
```

```python
import tradermade as tm
```

```python
# set api key
tm.set_rest_api_key(api_key)
```

```python
#get live data
tm.live(currency='AUDUSD',fields=["bid", "mid", "ask"]) # returns live data - fields is optional
```

```python
#get historical data

tm.historical(currency='AUDUSD', date="2011-01-20",interval="daily", fields=["open", "high", "low","close"]) 
# returns historical data for the currency requested interval is daily, hourly, minute - fields is optional
```

```python
# get timeseries data
import pandas as pd
df = pd.DataFrame()
for i in range(2012, 2022):  
    x = tm.timeseries(currency='AUDUSD', start=str(i)+"-12-01",fields=["open", "high", "low","close"], end=str(i+1)+"-11-30")
    df = df.append(x)
df = df.drop_duplicates()
df
# returns timeseries data for the currency requested interval is daily, hourly, minute - fields parameter is optional (you can select ["close"] if you just want close prices)
```

```python
df.index = pd.to_datetime(df['date'])
```

```python
df = df.drop(columns = "date")
df
```

```python
df.groupby(pd.Grouper(freq='M')).mean()
```
